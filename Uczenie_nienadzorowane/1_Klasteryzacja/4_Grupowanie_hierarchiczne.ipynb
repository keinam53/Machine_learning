### <a name='0'></a> Import bibliotek

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

sns.set(font_scale=1.2)
```

### <a name='1'></a> Wygenerowanie danych

```python
from sklearn.datasets import make_blobs

data = make_blobs(n_samples=10, centers=2, cluster_std=1, center_box=(-8, 8), random_state=42)[0]
df = pd.DataFrame(data, columns=['x1', 'x2'])
df
```

### <a name='2'></a> Wizualizacja danych

```python
plt.figure(figsize=(14, 7))
plt.scatter(data[:, 0], data[:, 1])

for label, x, y in zip(range(1, 11), data[:, 0], data[:, 1]):
    plt.annotate(s=label, xy=(x, y), xytext=(-3, 3), textcoords='offset points', ha='right', va='bottom')

plt.title('Grupowanie hierarhiczne')
plt.show()
```

### <a name='3'></a> Wizualizacja - dendrogram

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import linkage

#Przeprowadzanie grupowania hierarchicznego
linked = linkage(data)

#Wizualizacja grupowania hierarchicznego jako dendrogram
plt.figure(figsize=(14, 7))
dendrogram(linked, orientation='top', labels=range(1, 11), distance_sort='descending', show_leaf_counts=True)
plt.title('Grupowanie hierarchiczne - dendrogram')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
dendrogram(linked, orientation='right', labels=range(1, 11), distance_sort='descending', show_leaf_counts=True)
plt.title('Grupowanie hierarchiczne - dendrogram')
plt.show()
```

### <a name='4'></a> Grupowanie hierarchiczne

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=2)
cluster.fit_predict(data)
```

### <a name='5'></a> Wizualizacja klastrów

```python
df = pd.DataFrame(data, columns=['x1', 'x2'])
df['cluster'] = cluster.labels_
df.head()
```

```python
fig = px.scatter(df, 'x1', 'x2', color='cluster', width=950, height=500, template='plotly_dark', 
                 title='Grupowanie hierarchiczne', color_continuous_midpoint=0.6)
fig.update_traces(marker_size=12)
fig.show()
```

### <a name='6'></a> Porównanie metryk (euklidesowa, Manhattan, cosinusowa)

```python
data = make_blobs(n_samples=1000, centers=4, cluster_std=1.5, center_box=(-8, 8), random_state=42)[0]
df = pd.DataFrame(data, columns=['x1', 'x2'])

px.scatter(df, 'x1', 'x2', width=950, height=500, template='plotly_dark', 
                 title='Grupowanie hierarchiczne')
```

```python
#Odległość euklidesowa
cluster_euclidean = AgglomerativeClustering(n_clusters=4)
cluster_euclidean.fit_predict(data)
```

```python
df_euclidean = pd.DataFrame(data, columns=['x1', 'x2'])
df_euclidean['cluster'] = cluster_euclidean.labels_
df_euclidean.head()
```

```python
px.scatter(df_euclidean, 'x1', 'x2', color='cluster', width=950, height=500, template='plotly_dark', 
                 title='Grupowanie hierarchiczne - metryka euklidesowa')
```

```python
cluster_manhattan = AgglomerativeClustering(n_clusters=4, affinity='manhattan', linkage='complete')
cluster_manhattan.fit_predict(data)
```

```python
df_manhattan = pd.DataFrame(data, columns=['x1', 'x2'])
df_manhattan['cluster'] = cluster_manhattan.labels_
px.scatter(df_manhattan, 'x1', 'x2', color='cluster', width=950, height=500, template='plotly_dark', 
                 title='Grupowanie hierarchiczne - metryka manhattan')
```

```python
#Metryka cosinusowa
cluster_cosine = AgglomerativeClustering(n_clusters=4, affinity='cosine', linkage='complete')
cluster_cosine.fit_predict(data)
```

```python
df_cosine = pd.DataFrame(data, columns=['x1', 'x2'])
df_cosine['cluster'] = cluster_cosine.labels_
px.scatter(df_cosine, 'x1', 'x2', color='cluster', width=950, height=500, template='plotly_dark', 
                 title='Grupowanie hierarchiczne - metryka cosinusowa')
```

```python
from plotly.subplots import make_subplots

fig = make_subplots(rows=1, cols=3, column_titles=['euklidesowa', 'manhattan', 'cosinusowa'])

fig1 = px.scatter(df_euclidean, 'x1', 'x2', 'cluster')['data'][0]
fig2 = px.scatter(df_manhattan, 'x1', 'x2', 'cluster')['data'][0]
fig3 = px.scatter(df_cosine, 'x1', 'x2', 'cluster')['data'][0]

fig.add_trace(fig1, row=1, col=1)
fig.add_trace(fig2, row=1, col=2)
fig.add_trace(fig3, row=1, col=3)
fig.update_layout(template='plotly_dark')
fig.show()
```
