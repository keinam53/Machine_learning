### <a name='0'></a> Import bibliotek

```python
import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

### <a name='1'></a> Wczytanie danych

```python
data = pd.read_csv('https://storage.googleapis.com/esmartdata-courses-files/ml-course/factory.csv')
data.head()
```

```python
data.describe()
```

### <a name='2'></a> Wizualizacja danych

```python
px.scatter(data, 'item_length', 'item_width', width=950, template='plotly_dark', title='Isolation Forest')
```

### <a name='3'></a> Isolation Forset

```python
from sklearn.ensemble import IsolationForest

#contamination 0 - 0.05
outlier = IsolationForest(n_estimators=100, contamination=0.03)
outlier.fit(data)
```

```python
y_pred = outlier.predict(data)
y_pred[:30]
```

### <a name='4'></a> Isolation Forset - wizualizacja

```python
data['outlier_flag'] = y_pred
px.scatter(data, 'item_length', 'item_width', color='outlier_flag', width=950, 
           template='plotly_dark', title='Isolation Forest')
```
